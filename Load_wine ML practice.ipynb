```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
wine_df = pd.DataFrame(data=wine['data'],columns=wine['feature_names'])
wine_df['target'] = pd.Series(wine['target'])
wine_df
```

```python
X = wine_df.drop('target',axis=1)
y = wine_df['target']
X
```

```python
#Split into train and test
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#Choosing the best classification algorithm
```

```python
#import few classification algorithms
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
dict_class = {"Support Vector": SVC(),
              "KNN": KNeighborsClassifier(),
              "Logistic Regression": LogisticRegression(max_iter=600),
              "Decision Tree": DecisionTreeClassifier(),
              "Random Forest": RandomForestClassifier(n_estimators=30)}
model_test_score=[]
model_train_score=[]
for model_name, model in dict_class.items():
    model.fit(X_train, y_train)
    model_test_score.append(model_name + ":" + str(model.score(X_test, y_test)))
    model_train_score.append(model_name + ":" + str(model.score(X_train, y_train)))
print(model_test_score)
print(model_train_score)
```

```python
#k-fold cross validation
```

```python
X.shape
```

```python
Log_reg = LogisticRegression()
Rand_reg = RandomForestClassifier()
Log_reg.fit(X_train,y_train)
Rand_reg.fit(X_train,y_train)
from sklearn.model_selection import cross_val_score
print(cross_val_score(estimator=Log_reg,
                      X=X,
                      y=y,
                      cv=5,
                      verbose=2))
print(cross_val_score(estimator=Rand_reg,
                      X=X,
                      y=y,
                      cv=5,
                      verbose=2))
```

```python
Log_reg.get_params()
```

#Hyperparameter tuning

```python
from sklearn.model_selection import RandomizedSearchCV
grid = {'C': [1.0, 2.0],
        'max_iter': [100, 200, 500],
        'penalty': ['l1', 'l2', 'elasticnet'],
        'solver': ['liblinear','lbfgs']}
rs_logreg = RandomizedSearchCV(estimator=Log_reg,
                                param_distributions=grid,
                                n_iter=10,
                                cv=5,
                                verbose=1)
rs_logreg.fit(X_train,y_train)
print(rs_logreg.score(X_test, y_test))
print(rs_logreg.best_params_)
```

```python
np.logspace(-4, 4, 20)
```

```python
#Tuning Logistic regression parameters
from sklearn.model_selection import GridSearchCV
grid = {'C': np.logspace(-4,4,20),
        'max_iter': [50, 200, 500],
        'penalty': ['l1', 'l2', 'elasticnet'],
        'solver': ['liblinear','lbfgs']}
gs_logreg = GridSearchCV(estimator=Log_reg,
                          param_grid=grid,
                          cv=5,
                          verbose=2)
gs_logreg.fit(X_train, y_train)
print(gs_logreg.score(X_test, y_test))
print(gs_logreg.best_params_)
```

```python
Rand_reg.get_params()
```

#Hyperparameter tuning for Random forest

```python
from sklearn.model_selection import GridSearchCV
rand_grid = {'criterion': ['gini','entropy'],
             'max_features': ['auto','sqrt','log2'],
             'n_estimators': np.arange(10, 1000, 50)}
gs_randreg = GridSearchCV(estimator=Rand_reg,
                          param_grid=rand_grid,
                          cv=5,
                          verbose=2,
                          scoring=None)
gs_randreg.fit(X_train, y_train)
print(gs_randreg.score(X_test, y_test))
print(gs_randreg.best_params_)
```

```python
gs_logreg.best_params_, gs_randreg.best_params_
```

# Building a model with Logistic regression first

```python
log_reg_final = LogisticRegression(C=545.5594781168514,max_iter=50,penalty='l1',solver='liblinear')
log_reg_final.fit(X_train, y_train)
```

# Predicting Logistic Regression model

```python
log_pred = log_reg_final.predict(X_test)
```

# Visualizing true and predicted results

```python
plt.scatter(X_test['proline'],log_pred)
plt.scatter(X_test['proline'], y_test,c='red')
```

# Evaluating the model

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, log_pred))
print(classification_report(y_test, log_pred))
```

# Building a model with Random Forest's best parameters identified using Gridsearch

```python
rand_reg_final = RandomForestClassifier(max_features='log2')
rand_reg_final.fit(X_train, y_train)
```

# Predicting the model

```python
rand_pred = rand_reg_final.predict(X_test)
```

# Visualizing the true and predicted results

```python
plt.scatter(X_test['magnesium'],rand_pred)
plt.scatter(X_test['magnesium'],y_test,c='red')
```

# Evaluating the model

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test, rand_pred))
print(classification_report(y_test, rand_pred))
```

# Visualizing the confusion matrix

```python
fig, ax = plt.subplots(figsize=(3,2))
ax = sns.heatmap(confusion_matrix(y_test, rand_pred),
                 annot=True,
                 cmap = 'viridis')
plt.title('Confusion matrix')
plt.xlabel('Predicted label')
plt.ylabel('True label')
```
