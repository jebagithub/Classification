```python
# Separate out the categorical columns to scale the rest
categ_df = pot_haz_ds[['neo','pha','orbit_id','class']]
categ_df
```

```python
# Create a subset of only numerical colums
numer_df = pot_haz_ds.drop(['neo','pha','orbit_id','class'],axis=1)
numer_df.head()
```

```python
# Scale the numerical columns to prepare for machine learning
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_ar = scaler.fit_transform(numer_df)
scaled_ar
```

```python
# Convert the numpy array of scaled values to a dataframe
scaled_df = pd.DataFrame(scaled_ar,columns=numer_df.columns)
scaled_df
```

```python
categ_df.reset_index(inplace=True)
categ_df.drop('index',axis=1,inplace=True)
categ_df
```

```python
scaled_aster_df = pd.concat([scaled_df,categ_df],axis=1)
scaled_aster_df
```

```python
scaled_aster_df.isnull().sum()
```

```python
# Convert the categorical columns to numbers.
scaled_aster_final_df = pd.get_dummies(scaled_aster_df,columns=['neo','orbit_id','class'])
scaled_aster_final_df
```

```python
# Assign X and y and start testing various ML algorithms
X = scaled_aster_final_df.drop('pha',axis=1)
y = scaled_aster_final_df['pha']
```

```python
# My computer is not capable for training on such huge dataset. let me train on a sample.
X_sample = X[:100000]
y_sample=y[:100000]
```

```python
y_sample.value_counts()
```

```python
# Label encode the y values
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_sample = le.fit_transform(y_sample)
```

```python
unique, count = np.unique(y_sample,return_counts=True)
print(dict(zip(unique,count)))
```

```python
# Split train & test set from this sample
X_train_samp, X_test_samp, y_train_samp, y_test_samp = train_test_split(X_sample, y_sample, test_size=0.2, random_state=42)
```

```python
np.unique(y_test_samp,return_counts=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

model = {'Logistic Regression': LogisticRegression(),
         'Random Forest': RandomForestClassifier()}

model_score = {}
for model_name, model in model.items():
    model.fit(X_train_samp, y_train_samp)
    model_score[model_name] = model.score(X_test_samp, y_test_samp)
    
model_score
```

```python
# Let us try to capture the precision score, recall, accuracy and F1_score as well for both the algorithms.
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score

model = {'Logistic Regression': LogisticRegression(),
         'Random Forest': RandomForestClassifier()}
all_scores = {}
for model_name, model in model.items():
    model.fit(X_train_samp, y_train_samp)
    y_pred_samp = model.predict(X_test_samp)
    all_scores[model_name] = {"Precision Score": precision_score(y_test_samp, y_pred_samp),
                              "Recall Score": recall_score(y_test_samp, y_pred_samp),
                              "F1 score": f1_score(y_test_samp, y_pred_samp),
                              "Accuracy score": accuracy_score(y_test_samp,y_pred_samp)}
    
all_scores
```

```python
# It appears we may have overfit. Anyway, let us go ahead and use Random forest
from sklearn.ensemble import RandomForestClassifier

rf_clas_samp = RandomForestClassifier()
rf_clas_samp.fit(X_train_samp, y_train_samp)
rf_samp_pred = rf_clas_samp.predict(X_test_samp)

from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test_samp, rf_samp_pred)
print(cm)
```

```python
# Visualize the confusion matrix using a heatmap
plt.figure(figsize=(4,3))
sns.heatmap(cm,annot=True,cmap='Accent',fmt='.2f',cbar=False)
plt.xlabel('Predicted Label',fontsize=13)
plt.ylabel('True Label',fontsize=13)
```

```python
print(classification_report(y_test_samp, rf_samp_pred))
```
